```python
import pandas as pd
```

```python
#a)
def add_key(dict, key):
    if key not in dict:
        dict[key] = {
            "in": [],
            "out": []
        }
    return dict

def fill_dict(raw_data):
    graph = {}
    for from_node, to_node in zip(raw_data[" FromNodeId"], raw_data["ToNodeId"]):
        graph = add_key(graph, from_node)
        graph = add_key(graph, to_node)
        graph[from_node]["out"].append(to_node)
        graph[to_node]["in"].append(from_node)
    return graph

```

```python
path = "data/stanford_small.txt"
raw_data = pd.read_csv(path, delimiter="\t", header=0, comment="#")
print(raw_data.head(10))
print(len(graph))
graph = fill_dict(raw_data)
for i, (k, v) in enumerate(graph.items()):
    print(k, v)
    if i == 9:
        break
```

```python
#b)

def remove_dead_end(graph, dead_end):
        del graph[dead_end]
        for node in graph:
            if dead_end in graph[node]["out"]:
                graph[node]["out"].remove(dead_end)
        return graph

def dead_end_identifier(graph, input_nodes):
    for node in input_nodes:
        out_nodes = graph[node]["out"]
        in_nodes = graph[node]["in"]
        if not out_nodes:
            graph = remove_dead_end(graph, node)
            graph = dead_end_identifier(graph, in_nodes)
    return graph

graph = fill_dict(raw_data)

all_nodes = list(graph.keys())
graph_wo_dead_ends = dead_end_identifier(graph, all_nodes)
all_nodes_wo_dead_end = list(graph_wo_dead_ends.keys())
dead_ends = list(set(all_nodes) - set(all_nodes_wo_dead_end))
print(dead_ends)
print("Num Nodes Graph",len(all_nodes))
print("Num Nodes Graph wo Dead Ends", len(all_nodes_wo_dead_end))
print("Num Dead Ends", len(dead_ends))
```
